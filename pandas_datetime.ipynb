```python
# Timestamp → 문자열
```

```python
import pandas as pd

ts = pd.Timestamp('2024-06-04 09:30:45')

# strftime: Timestamp를 문자열로 변환

ts.strftime('%Y-%m-%d')           # '2024-06-04'
ts.strftime('%Y/%m/%d %H:%M:%S')  # '2024/06/04 09:30:45'
ts.strftime('%Y년 %m월 %d일')      # '2024년 06월 04일'
ts.strftime('%b %Y')              # 'Jun 2024'

# %b        Jan, Feb, Mar, ... Jun
# %B        January, February, March, ... June

# %Y        4자리 연도      2024
# %y        2자리 연도      24
# %m        2자리 월        06
# %B        월 전체 이름    June
# %b        월 약자        Jun
# %d        2자리 일        04
# %A        요일 전체 이름  Tuesday
# %a        요일 약자       Tue
# %H        24시간제 시     09
# %I        12시간제 시     09
# %p        AM/PM         AM
# %M        분             30
# %S        초             45
# %f        마이크로초      123456
# %j        연중 몇 번째 날 156
# %W        연중 몇 번째 주 23
```

```python
ts = pd.Timestamp.now()
# ts = pd.Timestamp('now')
print(ts.strftime('%Y-%m-%d'))
```

```python
ts1 = pd.to_datetime('2024/06/04')
ts2 = pd.to_datetime('20240604')
ts3 = pd.to_datetime('June 4, 2024')
ts4 = pd.to_datetime('2024-06-04 09:30:00')
ts5 = pd.to_datetime('04/06/2024', format='%d/%m/%Y')               # format으로 형식을 명시적으로 지정
ts6 = pd.to_datetime('2024년 06월 04일', format='%Y년 %m월 %d일')     # format으로 형식을 명시적으로 지정

print(ts1)
print(ts2)
print(ts3)
print(ts4)
print(ts5)
print(ts6)
```

```python
# pd.DatetimeIndex는 DataFrame/Series의 인덱스로 사용
```

```python
idx = pd.DatetimeIndex(['2024-01-01', '2024-06-04', '2024-12-31'])

print(idx)
print(idx.year)   # Index([2024, 2024, 2024])
print(idx.month)  # Index([1, 6, 12])
print(idx.day)    # Index([1, 4, 31])
```

```python
idx = pd.DatetimeIndex(['2024-01-01', '2024-06-04 09:30:45', '2024-12-31 23:59:59'])

print(idx.hour)    # Index([0, 9, 23], dtype='int32')
print(idx.minute)  # Index([0, 30, 59], dtype='int32')
print(idx.second)  # Index([0, 45, 59], dtype='int32')
```

```python
idx = pd.DatetimeIndex(['2024-01-01', '2024-06-04', '2024-12-31'])

print(idx.dayofweek)   # Index([0, 1, 1], dtype='int32')  0=월요일
print(idx.day_name())  # Index(['Monday', 'Tuesday', 'Tuesday'])
```

```python
idx = pd.DatetimeIndex(['2024-01-01', '2024-06-04', '2024-12-31'])

print(idx.quarter)  # Index([1, 2, 4], dtype='int32')
```

```python
idx = pd.DatetimeIndex(['2024-01-01', '2024-06-04', '2024-12-31'])

print(idx.dayofyear)  # Index([1, 156, 366], dtype='int32')
```

```python
# pd.date_range()의 리턴은 DatetimeIndex

idx = pd.date_range('2024-01-01', periods=12, freq='ME')
# pd.date_range()는 날짜 범위를 생성
# '2024-01-01'      시작 날짜
# periods=12        12개 생성
# freq='ME'         월말(Month End) 기준
# 'D': 일별, 'W': 주별, 'ME': 월말, 'QE': 분기말, 'YE': 연말

print(idx)
print(idx[0:6])  # 앞에서 6개  # 1월 ~ 6월 슬라이싱
```

```python
df = pd.DataFrame({
    '값': [10, 20, 30, 40, 50, 60]
}, index=pd.date_range('2024-01-01', periods=6, freq='ME'))

print(df)
```

```python
print(df['2024-01':'2024-03'])
```

```python
print(df.iloc[0:3])  # 정수 인덱스로 슬라이싱
```

```python
print(df.iloc[0])    # 첫 번째 행

# 값    10: 컬럼 이름이 "값"
# Name: 2024-01-31 00:00:00, dtype: int64
# Name은 인덱스
```

```python
idx = pd.date_range('2024-01-01', periods=3, freq='ME')
print(idx)
```

```python
print(idx.strftime('%Y-%m'))      # Index(['2024-01', '2024-02', '2024-03'])
print(idx.strftime('%Y년 %m월'))  # Index(['2024년 01월', '2024년 02월', '2024년 03월'])
```

```python
# pd.Timedelta
```

```python
print(pd.Timedelta('3 days'))
print(pd.Timedelta('3 days 2 hours 30 minutes'))
print(pd.Timedelta(days=3, hours=2, minutes=30))
```

```python
td = pd.Timedelta('3 days 2 hours 30 minutes')

print(td.days)         # 3

# seconds는 날짜를 제외한 나머지 시간 부분만 초로 반환
print(td.seconds)      # 9000  (2시간 30분 = 9000초)

# 날자를 포함한 전체 시간을 초로 바꾸려면 total_seconds()
print(td.total_seconds())  # 270600.0  (전체 초)
```

```python
ts = pd.Timestamp('2024-01-01')
td = pd.Timedelta('10 days')

print(ts + td)  # 2024-01-11 00:00:00
print(ts - td)  # 2023-12-22 00:00:00

# Timestamp끼리 빼면 Timedelta
print(pd.Timestamp('2024-06-04') - pd.Timestamp('2024-01-01'))  # 155 days
```

```python
td1 = pd.Timedelta('3 days')
td2 = pd.Timedelta('5 days')

print(td1 > td2)   # False
print(td1 < td2)   # True
print(td1 == td2)  # False
```
