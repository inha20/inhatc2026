```python
# px.histogram

fig = px.histogram(df_tips, x="total_bill",
                   color="sex",
                   nbins=30,
                   # histnorm="probability",
                   title="성별 총 계산서 분포",
                   # hover_data=["smoker"], 
                   # histogram은 개별 데이터 포인트가 아니라 bin으로 집계된 데이터를 표시하기 때문에 표시 안됨
                   labels={"total_bill": "총 계산서", "sex": "성별"})

# hover에 표시되는 대상은 데이터프레임 컬럼을 전달받는 인자: x, color
# y는 넘겨주지 않았지만 histnorm의 값이 표시
# title, labels, nbins, histnorm — 차트의 표현 방식을 바꾸는 인자 → hover에 미표시

# histnorm의 값
# None                      기본값, bin 안의 데이터 개수(count) 표시
# "percent"                 전체 대비 각 bin의 비율을 % 로 표시
# "probability"             전체 대비 각 bin의 비율을 0~1 사이 값으로 표시
# "density"                 bin의 크기로 나눈 밀도값 표시
# "probability density"     모든 bin의 넓이 합이 1이 되도록 정규화
fig.show()
```

```python
# go.Sankey
# 해당 데이터 시각화 방법을 만들어 낸 사람 이름
# 증기 엔진의 에너지 흐름을 시각화 (1898)

fig = go.Figure(go.Sankey(
    node=dict(
        label=["A", "B", "C", "D", "E"],
        color=["blue", "green", "red", "orange", "purple"]
    ),
    link=dict(
        source=[0, 0, 1, 2, 3],
        target=[2, 3, 3, 4, 4],
        value=[8, 4, 2, 8, 4]
    )
))

fig.update_layout(title="Sankey 다이어그램")
fig.show()


# node — 차트에 표시되는 박스(노드)를 정의. 이름과 색상 등 노드 자체의 속성을 설정
# link — 노드 간의 연결과 흐름을 정의. 어디서 어디로 얼마만큼 흐르는지를 설정

# source                  target                  value
# A(0)                    C(2)                    8
# A(0)                    D(3)                    4
# B(1)                    D(3)                    2
# C(2)                    E(4)                    8
# D(3)                    E(4)                    4
```

```python
df_wind = px.data.wind()
df_wind.dtypes

# direction                 풍향 (N, NNE, NE 등 16방위)
# strength                  풍속 구간 (0-1, 1-2 등)
# frequency                 해당 풍향/풍속의 발생 빈도
```

```python
df_wind.head()
```

```python
# go.Scatterpolar

fig = go.Figure(go.Scatterpolar(
    r=df_wind["frequency"],
    theta=df_wind["direction"],
    mode="markers",
    name="풍향/풍속 빈도"
))

fig.update_layout(title="풍향별 빈도")
fig.update_layout(
    # x, y가 아닌 원점에서의 거리와 각도로 표현하는 극좌표
    # polar는 극좌표 차트의 레이아웃을 설정하는 파라미터
    polar=dict(
        # angularaxis는 방향, 회전 등을 설정
        angularaxis=dict(
            direction="counterclockwise",   # N → E → S → W를 시계방향/반시계방향으로 배치할지를 결정
            rotation=90                     # clockwise: 90이 디폴트 (12시 방향), counterclockwise: 0이 디폴트 (3시 방향) 
        )
    )
)
fig.show()

# r        반지름 값 (중심에서의 거리)
# theta    각도 값 (방향)
# mode     lines, markers, lines+markers 중 선택
```

```python
# go.Barpolar

fig = go.Figure(go.Barpolar(
    r=df_wind["frequency"],
    theta=df_wind["direction"],
    name="풍향/풍속 빈도"
))

fig.update_layout(title="풍향별 빈도")
fig.update_layout(
    # x, y가 아닌 원점에서의 거리와 각도로 표현하는 극좌표
    # polar는 극좌표 차트의 레이아웃을 설정하는 파라미터
    polar=dict(
        # angularaxis는 방향, 회전 등을 설정
        angularaxis=dict(
            direction="clockwise",          # N → E → S → W를 시계방향/반시계방향으로 배치할지를 결정
            rotation=90                     # clockwise: 90이 디폴트 (12시 방향), counterclockwise: 0이 디폴트 (3시 방향) 
        )
    )
)
fig.show()

# r        막대의 길이 (중심에서의 거리)
# theta    막대의 방향 (각도)
```

```python
# go.Candlestick

# python -m pip install yfinance

import yfinance as yf

df_aapl = yf.download("AAPL", start="2024-01-01", end="2024-12-31")
df_aapl.head()
```

```python
# go.Candlestick

fig = go.Figure()

fig.add_trace(go.Candlestick(
    x=df_aapl.index,
    open=df_aapl["Open"]["AAPL"],       # 멀티인덱스
    high=df_aapl["High"]["AAPL"],
    low=df_aapl["Low"]["AAPL"],
    close=df_aapl["Close"]["AAPL"],
    name="AAPL",                        # trace 이름이며 legend에 표시
    hovertext="AAPL 주가",
    # hovertext는 x, open, high, low, close, name 외에 추가로 보여주고 싶은 텍스트를 지정

    hoverinfo="x+text"
    # hoverinfo는 디폴트가 all
    # 보여줄 대상을 지정하는 방법이 hoverinfo; text는 hovertext를 의미
))

fig.add_trace(go.Scatter(
    x=df_aapl.index,
    y=df_aapl["Close"]["AAPL"].rolling(20).mean(),
    mode="lines",
    name="20일 이동평균"
))

fig.update_layout(title="AAPL 캔들스틱 + 20일 이동평균선",
                  xaxis_title="날짜",
                  yaxis_title="가격(USD)")
fig.show()
```
